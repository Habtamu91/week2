```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='rating', hue='bank')
plt.title('Rating Distribution by Bank')
plt.show()

```

```python
df['review_length'] = df['review'].apply(len)
sns.histplot(data=df, x='review_length', bins=30, hue='bank', kde=True)
plt.title('Review Length Distribution')
plt.show()
```

```python
sns.boxplot(data=df, x='rating', y='review_length')
plt.title('Review Length by Rating')
plt.show()
```

```python
sns.countplot(data=df, x='bank')
plt.title('Number of Reviews per Bank')
plt.show()
```

```python
df['date'] = pd.to_datetime(df['date'])
df.groupby(df['date'].dt.to_period('M')).size().plot()
plt.title('Number of Reviews Over Time')
plt.show()
```
